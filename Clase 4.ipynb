```python
import math as m
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2*np.pi,2*np.pi,200)
w = [1,np.pi,15,30]
for w_i in w:
    y = np.sinc(w_i*x)
    plt.plot(x,y,label=r"${\rm sinc (%2.1f x)}$" %w_i)
plt.title(r"${\rm sinc (\omega x)}$")
plt.legend()
plt.show()
```
