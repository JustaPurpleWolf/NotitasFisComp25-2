```python
import math as m
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0.5,5,200)
y = -2/x**6 + 1/x**12
```

```python
plt.plot(x,y)
plt.xlim([x[0],x[-1]]),plt.ylim(-1.1,1.1)
plt.grid()
plt.show()
```

```python
plt.loglog(x,y+1)
plt.grid()
plt.show()
```

```python
dy = 12/x**7 - 12/x**13
```

```python
plt.plot(x,dy)
plt.xlim([x[0],x[-1]]),plt.ylim(-4,4)
plt.grid()
plt.show()
```

```python
plt.loglog(x,-dy+3)
plt.loglog(x,-dy+3,'.')
plt.grid()
plt.show()
```
