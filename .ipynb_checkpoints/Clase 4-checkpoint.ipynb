```python
import math as m
import numpy as np
import matplotlib.pyplot as plt
```

$$
{\rm sinc} (\omega x)
$$

```python
x = np.linspace(-2*np.pi,2*np.pi,200)
w = [1,np.pi,15,30]
for w_i in w:
    y = np.sinc(w_i*x)
    plt.plot(x,y,label=r"${\rm sinc} (%2.1f x)$" %w_i)
plt.title(r"${\rm sinc} (\omega x)$")
plt.legend()
plt.show()
```

```python
x = np.linspace(-2*np.pi,2*np.pi,200)
w = [1,np.pi,15,30]
for i in range(len(w)):
    y = np.sinc(w[i]*x)
    plt.plot(x,y,label=r"${\rm sinc} (%2.1f x)$" %w[i])
plt.title(r"${\rm sinc} (\omega x)$")
plt.legend()
plt.show()
```

```python
def Heaviside(x,x0=0):
    if x < x0:
        H = 0
    elif x == x0:
        H = 0.5
    else:
        H = 1
    return H

def Heaviside_array(x,x0=0):
    H = np.zeros_like(x)
    for i in range(len(x)):
        H[i] = Heaviside(x[i],x0)
    return H

plt.plot(x,Heaviside_array(x,2))
plt.show()
```
