```python
import math as m
import random as rd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Vcuadcuad(x,y,L=0.5,V0=1.0):
    Nx , Ny = len(x) , len(y)
    malla = np.zeros((Nx,Ny))
    for i in range(1,Nx-1):
        for j in range(1,Ny-1):
            if  -0.5*L < x[i] < 0.5*L and -0.5*L < y[j] < 0.5*L:
                malla[i,j] = 1
            else:
                malla[i,j] = rd.random()
    return V0*malla

def Relajar(V,V0=1):
    Nx , Ny = V.shape
    for i in range(1,Nx-1):
        for j in range(1,Ny-1):
            if V[i,j] != 1:
                V[i,j] = 0.25*( V[i+1,j] + V[i-1,j] + V[i,j+1] + V[i,j-1] )
    return V
```

```python
x , y = np.linspace(-2,2,100) , np.linspace(-2,2,100)
pot = Vcuadcuad(x,y)
```

```python
for i in range(1000):
    pot = Relajar(pot)
plt.imshow(pot)
plt.show()
```

```python
plt.contour(pot)
plt.gca().set_aspect('equal')
plt.show()
```
