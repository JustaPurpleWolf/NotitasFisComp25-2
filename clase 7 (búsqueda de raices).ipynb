```python
import math as m
import numpy as np
import scipy.special as sp
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5,5,100)
y = sp.airy(x)[0]
plt.plot(x,y)
plt.title("Funcionalistas de Airy")
plt.grid()
plt.show()
```

```python
def mi_airy(x):
    return sp.airy(x)[0]
```

```python
def buscar_raiz(f,x_izq,x_der):
    f_izq = f(x_izq)
    for i in range(10):
        x_med = 0.5*(x_izq+x_der)
        f_med = f(x_med)
        fifm = f_izq*f_med
        if fifm < 0:
            x_izq , x_der = x_izq , x_med
        else:
            x_izq , x_der = x_med , x_der
            f_izq = f_med
        print('Valor de f = ',f_med,'\t Propuesta de raiz = ',x_med)
    return x_med
```

```python
x0 = buscar_raiz(mi_airy,-5,-4)
```

```python
x = np.linspace(-5,5,100)
y = sp.airy(x)[0]
plt.title("Funcionalistas de Airy")
plt.plot(x,y,label=r"$Ai(x)$")
plt.plot(x0,[0],'x',label=r"raíz $x_0$")
plt.grid(),plt.legend()
plt.show()
```
